```python
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars = pd.read_csv('imports-85.data', names=cols)
```

```python
cols = [ 'normalized-losses',
         'wheel-base', 'length','width', 
        'height', 'curb-weight','bore','stroke',
        'compression-rate','horsepower',
        'peak-rpm','city-mpg','highway-mpg','price']

num_cars = cars[cols]
num_cars
```

```python
num_cars = num_cars.replace('?',np.nan)
```

```python
num_cars = num_cars.astype(float)
```

```python
num_cars.isnull().sum()
```

```python
num_cars = num_cars.fillna(num_cars.mean())
```

```python
num_cars
```

```python
price = num_cars['price']
num_cars = (num_cars - num_cars.min())/(num_cars.max() - num_cars.min())
num_cars['price'] = price
```

```python
num_cars
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
def knn_train_test(train_col,target_col,df):
    knn = KNeighborsRegressor()
    np.random.seed(1)
    index = np.random.permutation(df.index)
    df = df.reindex(index)
    
    num = int(len(df)/2)
    train = df.iloc[:num]
    test = df.iloc[num:]
    
    knn.fit(train[train_col],train[target_col])
    pre = knn.predict(test[train_col])
    mse = mean_squared_error(pre,test[target_col])
    rmse = mse**(1/2)
    return rmse    
```

```python
def knn_train_testV2(train_col,target_col,df,k):
    knn = KNeighborsRegressor(n_neighbors = k)
    np.random.seed(1)
    index = np.random.permutation(df.index)
    df = df.reindex(index)
    
    num = int(len(df)/2)
    train = df.iloc[:num]
    test = df.iloc[num:]
    
    knn.fit(train[train_col],train[target_col])
    pre = knn.predict(test[train_col])
    mse = mean_squared_error(pre,test[target_col])
    rmse = mse**(1/2)
    return rmse   
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
k1 = {}
k3 = {}
k5 = {}
k7 = {}
k9 = {}
for v in cols:
    test1 = knn_train_testV2([v],['price'],num_cars,1)
    test2 = knn_train_testV2([v],['price'],num_cars,3)
    test3 = knn_train_testV2([v],['price'],num_cars,5)
    test4 = knn_train_testV2([v],['price'],num_cars,7)
    test5 = knn_train_testV2([v],['price'],num_cars,9)
    t = [test1,test2,test3,test4,test5]
    x = range(len(t))
    plt.scatter(x,t)
    plt.show()
    k1[v] = test1
    k3[v] = test2
    k5[v] = test3
    k7[v] = test4
    k9[v] = test5
```
